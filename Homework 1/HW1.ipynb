```python
import math
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("glasshw1.csv", names=['RI','Na','Mg','Al','Si','K','Ca','Ba','Fe','Glass'], index_col=0)
df.reset_index(drop=True, inplace=True)
df
```

```python
total = df.shape[0]
print("Total observations = ", total)
c1_count,c2_count = pd.value_counts(df['Glass'].values, sort=False)
print("Class 1 count = ", c1_count)
print("Class 2 count = ", c2_count)
```

```python
c1_df = df[df['Glass'] == 1]
c2_df = df[df['Glass'] == 2]
```

```python
c1_df.reset_index(drop=True, inplace=True)
c2_df.reset_index(drop=True, inplace=True)
```

```python
col_names = list(df)
col_names.remove('Glass')

mean_c1 = dict.fromkeys(col_names,0)
sd_c1 = dict.fromkeys(col_names,0)

mean_c2 = dict.fromkeys(col_names,0)
sd_c2 = dict.fromkeys(col_names,0)
```

```python
for index,row in c1_df.iterrows():
    for i in col_names:
        mean_c1[i]+=row[i]

for i in col_names:
        mean_c1[i]/=c1_count

for index,row in c1_df.iterrows():
    for i in col_names:
        sd_c1[i]+=(row[i]-mean_c1[i])**2

for i in col_names:
    sd_c1[i]=math.sqrt(sd_c1[i]/(c1_count-1))
```

```python
for index,row in c2_df.iterrows():
    for i in col_names:
        mean_c2[i]+=row[i]

for i in col_names:
        mean_c2[i]/=c2_count

for index,row in c2_df.iterrows():
    for i in col_names:
        sd_c2[i]+=(row[i]-mean_c2[i])**2

for i in col_names:
    sd_c2[i]=math.sqrt(sd_c2[i]/(c2_count-1))
```

```python
def calc_gaussian(mean, sd, xi):
    return (math.exp(-((xi-mean)**2)/(2*sd*sd)))/(math.sqrt(2*math.pi)*sd)
```

```python
p_c1 = c1_count/total
p_c2 = c2_count/total
```

```python
log_sum_list=[]

for index,row in df.iterrows():
    temp_c1=0
    temp_c2=0
    for i in col_names:
        temp_c1+=math.log(calc_gaussian(mean_c1[i], sd_c1[i], row[i]))
        temp_c2+=math.log(calc_gaussian(mean_c2[i], sd_c2[i], row[i]))
    temp_c1+=math.log(p_c1)
    temp_c2+=math.log(p_c2)
    if temp_c1 > temp_c2:
        log_sum_list.append(int(1))
    elif temp_c1 < temp_c2:
        log_sum_list.append(int(2))
    elif temp_c1 == temp_c2:
        if c1_count > c2_count:
            log_sum_list.append(int(1))
        else:
            log_sum_list.append(int(2))

log_sum_series = pd.Series(log_sum_list)
log_sum_series = log_sum_series.astype('int64')
df.insert(loc=len(df.columns),column='Prediction', value=log_sum_series)
df
```

```python
count=0
for index,row in df.iterrows():
    if row['Glass'] != row['Prediction']:
        count+=1
training_error = count/total
print("Training Error = {0} %".format(training_error*100))
```
